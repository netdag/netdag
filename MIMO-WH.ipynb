```python
import netdag
import network_statistics
from IPython.display import Image
from numpy import zeros
import matplotlib.pyplot as plt
from tqdm import tqdm
from itertools import product
from time import sleep

models = {}
WH = [(19, 20), (19, 40), (15, 20), (15, 40), (15, 60), (15, 80), (10, 20), (5, 20)]
g = netdag.load_graph('control-application.xml.gz')
actuators = [int(v) for v in g.vertices() if len(g.get_out_neighbors(v)) == 0][:-1]
```

```python
Image(filename='control-application.xml.gz_task-graph.png')
```

```python
for actuator_count, WH_i in tqdm(product(range(1, len(actuators)+1), WH), total=len(actuators)*len(WH)):
    g = netdag.load_graph('control-application.xml.gz')
    if (actuator_count, WH_i) in models:
        continue
    for actuator in actuators[:actuator_count]:
        g.vertex_properties['weakly-hard'][actuator] = WH_i
    network = {'A': 1, 'B': 2, 'C': 3, 'D': 1, 'GAMMA': 1,
               'LAMBDA': network_statistics.MIMO_lambda_weakly_hard}
    model = netdag.get_makespan_optimal_weakly_hard_schedule(g,
                                                             network,
                                                             feasibility_timeout=30*60*1000,
                                                             optimization_timeout=20*60*1000)
    zeta, chi, duration, label = model
    models[(actuator_count, WH_i)] = model
```

```python
for WH_i in WH:
    print(WH_i)
    for actuator_count in range(1, len(actuators)+1):
        print(
            '\t%d -> %s' % (
                actuator_count,
                str(max(models[(actuator_count, WH_i)][0])) if models[(actuator_count, WH_i)][0] else 'infeasible'))
    
```
