```python
import netdag
import network_statistics
from IPython.display import Image
from numpy import zeros
import matplotlib.pyplot as plt
from tqdm import tqdm

models = {}
WH = [(19, 20), (19, 40), (15, 20), (15, 40), (15, 60), (15, 80)]
```

```python
Image(filename='control-application.xml.gz_task-graph.png')
```

```python
g = netdag.load_graph('control-application.xml.gz')
makespan = zeros(len(WH))
assert(netdag.is_DAG(g))
for i, WH_i in tqdm(list(enumerate(WH))):
    if WH_i in models:
        print('found model for', WH_i, '(skipping)')
        continue
    for v in g.vertices():
        if g.vertex_properties['soft'][v] >= 0:
            g.vertex_properties['weakly-hard'][v] = WH_i
    network = {'A': 1, 'B': 2, 'C': 1, 'D': 1, 'GAMMA': 1,
               'LAMBDA': network_statistics.MIMO_lambda_weakly_hard}
    model = netdag.get_makespan_optimal_weakly_hard_schedule(g, network)
    zeta, chi, duration, label = model
    models[WH_i] = model
    makespan[i] = max(zeta) if zeta is not None else float('inf')
```

```python
print(list(zip(WH, makespan)))
```
