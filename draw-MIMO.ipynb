```python
from graph_tool import load_graph
from graph_tool.draw import graph_draw, sfdp_layout
from IPython.display import Image, display

g = load_graph('control-application.xml.gz')

pos = sfdp_layout(g)
pos[0] = [0, 0]
pos[1] = [0, 1*4/6]
pos[2] = [0, 2*4/6]
pos[3] = [0, 3*4/6]
pos[4] = [0, 4*4/6]
pos[5] = [0, 5*4/6]
pos[6] = [2, 1-.3]
pos[7] = [2, 2-.3]
pos[8] = [2, 3-.3]
pos[9] = [4, 0]
pos[10] = [4, 1]
pos[11] = [4, 2]
pos[12] = [4, 3]


fill_color = g.new_vertex_property('vector<float>')
for v, color in zip(g.vertices(), [[.89,.902,.918,1]]*6+[[.718,.749,.796,1]]*3+[[.529,.58,.659,1]]*4):
    fill_color[v] = color


graph_draw(
    g,
    pos=pos,
    output='control-application.xml.gz_task-graph.png',
    vertex_text=g.vertex_index,
    vertex_font_size=18,
    vertex_fill_color=fill_color,
    vertex_color=[.251,.337,.471,1],
    edge_color=[.251,.337,.471,1],
    edge_pen_width=4,
    edge_marker_size=20
)

Image(filename='control-application.xml.gz_task-graph.png')
```
